# Общая информация

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Общие требования к программе:

* код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
* в коде отсутствуют глобальные переменные (за исключением documents и directories);
* пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

```python
#перечень документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

#перечень полок
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Задание 1

def doc_owher(document, num): #Пункт 1  - найти владельца документа по номеру
    for doc in document:
        if doc.get('number') == num:
            return doc.get('name')
    return 'Документ не найден в базе'

def number_directory(directory, number): #Пункт 2 - найти на какой полке документ по номеру
    for key, value in directory.items():
        if number in value:
            return('Документ хранится на полке: '+key)
    return 'Документ не найден в базе'
   

def full_info(document): #Пункт 3 - вывести все документы с полками
    for doc in document:
        print('№:'+ doc.get('number')+', тип: '+ doc.get('type')+', владелец: '+ doc.get('name')+ ' - ' +number_directory(directories, doc.get('number')))
    
        
def direct_exist(directory, num_dir): # Полка существует
    for key in directory:
        if key == num_dir:
            return True
    return False
        

def add_directory (directory, num_dir): #Пункт 4 - добавить полку
    if direct_exist(directory, num_dir) == False:
        directory [num_dir] = []
        print('Полка добавлена. Текущий список полок: ')
    else: print('Такая полка уже существует. Текущий список полок: ')
    for key in directory.keys():
        print(key)


def direct_empty(directory, num_dir): # Полка пустая
    for key, value in directory.items():
        if (key == num_dir) and (len(value) > 0):
            return False
    return True
        
def del_direct(directory, num_dir): #Пункт 5 - удалить полку
    if direct_exist(directory, num_dir) == True:
        if direct_empty(directory, num_dir) == True:
            del directory[num_dir]
            print('Полка удалена. Текущий список полок: ')
        else: print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ')
    else: print('Такой полки не существует. Текущий список полок: ')
    
    for key in directory.keys():
        print(key)

#Взаимодействие с пользователем        
while True:
    user_command = input("Введите команду: ")
    if user_command == 'p': 
        num = input('Введите номер документа: ')
        print(doc_owher(documents, num))
    elif user_command == 's':
        number = input('Введите номер документа: ')
        print(number_directory(directories, number))
    elif user_command == '/':
        print(full_info(documents))
    elif user_command == 'ads':
        num_directory = input('Введите номер полки: ')
        print(add_directory(directories, num_directory))
    elif user_command == 'ds':
        num_directory = input('Введите номер полки: ')
        print(del_direct(directories, num_directory))
    elif user_command == 'q': 
        print('Выполнение завершено')
        break
    else: print('Такой команды не существует')
```
